```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
#Review data
df_market_data.info()
```

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

---

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
crypto_data_scaled = StandardScaler().fit_transform(df_market_data)
```

```python
# Create a DataFrame with the scaled data
df_crypto_transformed = pd.DataFrame(crypto_data_scaled, columns=df_market_data.columns)
```

```python
# Copy the crypto names from the original data
df_crypto_transformed["coin_id"] = df_market_data.index
```

```python
# Set the coinid column as index
df_crypto_transformed = df_crypto_transformed.set_index("coin_id")


# Display sample data
df_crypto_transformed.head(10)
```

---

### Find the Best Value for k Using the Original Data.

```python
# Create a list with the number of k-values from 1 to 11
k_values = list(range(1,12))
k_values
```

```python
# Create an empty list to store the inertia values
inertia_values = []
```

```python
# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_scaled`
# 3. Append the model.inertia_ to the inertia list
for k in k_values:
    k_model = KMeans(n_clusters=k, random_state=1, n_init="auto")
    k_model.fit(df_crypto_transformed)
    inertia_values.append(k_model.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_curve = {"k": k_values, "inertia": inertia_values}
```

```python
# Create a DataFrame with the data to plot the Elbow curve
df_elbow_curve = pd.DataFrame(elbow_curve)
df_elbow_curve
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
elbow_plot = df_elbow_curve.hvplot.line(
    x='k',
    y='inertia',
    title='Elbow Curve',
    xticks=k_values,
    value_label='Original Elbow Curve'
)

elbow_plot
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** 5 due the clear and distinct elbow from 5 in which the changes in inertia is fairly flat.  

---

### Cluster Cryptocurrencies with K-means Using the Original Data

```python
# Initialize the K-Means model using the best value for k
k_value = 5
crypto_model = KMeans(n_clusters=k_value, random_state=1, n_init='auto')
```

```python
# Fit the K-Means model using the scaled data
crypto_model.fit(df_crypto_transformed)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled data
k_crypto_groups = crypto_model.predict(df_crypto_transformed)
```

```python
# Print the resulting array of cluster values
k_crypto_groups
```

```python
# Create a copy of the DataFrame
df_crypto_data_predictions = df_crypto_transformed.copy()
```

```python
# Add a new column to the DataFrame with the predicted clusters
df_crypto_data_predictions['predicted_cluster'] = k_crypto_groups

# Display sample data
df_crypto_data_predictions.head(10)
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.

crypto_scatter_plot = df_crypto_data_predictions.hvplot.scatter(
    x="price_change_percentage_24h",
    y="price_change_percentage_7d",
    by="predicted_cluster",
    hover_cols="coin_id",
    title="Predict Crypto Market Clusters"
)

crypto_scatter_plot
```

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
pca = PCA(n_components=3)
```

```python
# Use the PCA model with `fit_transform` to reduce to 
# three principal components.
crypto_pca = pca.fit_transform(df_crypto_transformed)


# View the first five rows of the DataFrame. 
crypto_pca[:5]
```

```python
# Retrieve the explained variance to determine how much information 
# can be attributed to each principal component.
pca.explained_variance_ratio_
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:** 88% of total variance is condensed into the three PCA variables. 

```python
# Create a new DataFrame with the PCA data.
df_crypto_pca = pd.DataFrame(
    crypto_pca,
    columns=['PCA1', 'PCA2', 'PCA3']
)
```

```python
# Copy the crypto names from the original data
df_crypto_pca['coin_id'] = df_market_data.index
```

```python
# Set the coinid column as index
df_crypto_pca = df_crypto_pca.set_index('coin_id')

# Display sample data
df_crypto_pca
```

---

### Find the Best Value for k Using the PCA Data

```python
# Create a list with the number of k-values from 1 to 11
k_values = list(range(1,12))
k_values
```

```python
# Create an empty list to store the inertia values
inertia_pca = []
```

```python
# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_pca`
# 3. Append the model.inertia_ to the inertia list

for k in k_values:
    k_model = KMeans(n_clusters=k, random_state=1, n_init='auto')
    k_model.fit(df_crypto_pca)
    inertia_pca.append(k_model.inertia_)

inertia_pca
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_pca = {
    'k': k_values,
    'inertia': inertia_pca
}
```

```python
# Create a DataFrame with the data to plot the Elbow curve
df_elbow_crypto_pca = pd.DataFrame(elbow_pca)
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.

elbow_pca_plot = df_elbow_crypto_pca.hvplot.line(
    x='k',
    y='inertia',
    title="Elbow Curve using PCA",
    xticks=k_values,
    value_label='PCA Elbow Curve'
)

elbow_pca_plot
```

#### Answer the following questions: 

* **Question:** What is the best value for `k` when using the PCA data?

  * **Answer:** 4 because of the steep drop between 2 and 4. 


* **Question:** Does it differ from the best k value found using the original data?

  * **Answer:** There are less distinct kinks between the data as in the original data elbow.

### Cluster Cryptocurrencies with K-means Using the PCA Data

```python
# Initialize the K-Means model using the best value for k
k_bestvalue_pca = 4
model_pca = KMeans(n_clusters=k_bestvalue_pca,
                   random_state=1, n_init='auto')
```

```python
# Fit the K-Means model using the PCA data
model_pca.fit(df_crypto_pca)
```

```python
# Predict the clusters to group the cryptocurrencies using the PCA data
k_crypto_groups_pca = model_pca.predict(df_crypto_pca)

# Print the resulting array of cluster values.
k_crypto_groups_pca
```

```python
# Create a copy of the DataFrame with the PCA data
df_crypto_pca_prediction = df_crypto_pca.copy()


# Add a new column to the DataFrame with the predicted clusters
df_crypto_pca_prediction['predicted_cluster'] = k_crypto_groups_pca


# Display sample data
df_crypto_pca_prediction
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="PC1"` and `y="PC2"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.

crypto_pca_prediction_plot = df_crypto_pca_prediction.hvplot.scatter(
    x='PCA1', 
    y='PCA2',
    by='predicted_cluster',
    hover_cols='coin_id',
    title="Predicted Crypto Market Clusters with PCA"
)

crypto_pca_prediction_plot
```

### Visualize and Compare the Results

In this section, you will visually analyze the cluster analysis results by contrasting the outcome with and without using the optimization techniques.

```python
# Composite plot to contrast the Elbow curves
elbow_plots_combined = elbow_plot * elbow_pca_plot

elbow_plots_combined.opts(
    title='Elbow Plots Comparison'
)

display(elbow_plots_combined)
```

```python
# Composite plot to contrast the clusters
elbow_clusters_combined = crypto_scatter_plot + crypto_pca_prediction_plot

elbow_clusters_combined.opts(
    title="Crypto Cluster Comparison: K Means vs. PCA"
)

display(elbow_clusters_combined)
```
